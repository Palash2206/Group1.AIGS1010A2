```python
import numpy as np
from matplotlib import pyplot as plt
import cv2
import numpy as np
import os
%matplotlib inline
```

```python

#Q1
colored_image = cv2.imread('takken.png',1) # reading as color
grayscale_image = cv2.imread('takken.png',0) # reading as grayscale
unchanged_image = cv2.imread('takken.png')
```

```python
plt.imshow(colored_image)
```

```python
sobelx = cv2.Sobel(grayscale_image, cv2.CV_64F, 1, 0, ksize=3)
sobely = cv2.Sobel(grayscale_image, cv2.CV_64F, 0, 1, ksize=3)
```

```python
#Q1a and Q1b To create a plot with two subplots, one for Gx and one for Gy
plt.subplot(1, 2, 1)
plt.imshow(sobelx, cmap='gray')
plt.title('Sobel X')

plt.subplot(1, 2, 2)
plt.imshow(sobely, cmap='gray')
plt.title('Sobel Y')

plt.show()
```

```python
#Q1c To create Gradient magnitude G 
grad_mag = np.sqrt(sobelx**2 + sobely**2)
```

```python
plt.imshow(grad_mag, cmap='gray')
plt.title('Gradient Magnitude')
plt.show()
```

```python
#Q2a
import urllib.request

# Load the image and add salt and pepper noise
img = cv2.imread("takken.png")
noise = np.zeros_like(img)
cv2.randu(noise, 0, 255)
img[noise < 10] = 0
img[noise > 245] = 255

# Display the noisy image
cv2.imshow("Noisy Image", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Q2b

# Apply median filter with a window size of 3x3
img_filtered1 = cv2.medianBlur(img, 3)

# Apply median filter with a window size of 5x5
img_filtered2 = cv2.medianBlur(img, 5)

# Display the original and filtered images
cv2.imshow("Original Image", img)
cv2.imshow("Filtered Image (3x3)", img_filtered1)
cv2.imshow("Filtered Image (5x5)", img_filtered2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Q2c

# Apply Gaussian filter with a kernel size of 5x5 and standard deviation of 0.5
img_filtered = cv2.GaussianBlur(img, (5, 5), 0.5)

# Display the original and filtered images
cv2.imshow("Original Image", img)
cv2.imshow("Filtered Image", img_filtered)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#2d

# Apply median filter with a window size of 3x3
img_median = cv2.medianBlur(img, 3)

# Apply Gaussian filter with a kernel size of 5x5 and standard deviation of 0.5
img_gaussian = cv2.GaussianBlur(img, (5, 5), 0.5)

# Display the original and filtered images
cv2.imshow("Original Image", img)
cv2.imshow("Median Filtered Image", img_median)
cv2.imshow("Gaussian Filtered Image", img_gaussian)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Load the image
img = cv2.imread('takken.png', 0)

# Define the range of scale and contrast thresholds to test
scale_thresholds = [30, 60, 90]
contrast_thresholds = [0.1, 0.2]

# Apply Canny edge detection with varying thresholds and display the results
for scale in scale_thresholds:
    for contrast in contrast_thresholds:
        edges = cv2.Canny(img, contrast * np.mean(img), contrast * np.mean(img) + scale)
        cv2.imshow(f'Scale: {scale}, Contrast: {contrast}', edges)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
